A pesar de que previamente se había realizado un proceso de preprocesamiento en el entorno de SQL, se opta por abordar una parte del preprocesamiento en el entorno de Python.
```python
#### crear base de datos para manejo de datos ####
conn = sql.connect("/content/drive/MyDrive/trabajo/Trabajo-analitica-RH/data/bd_empleados.csv") ### crea una base de datos con el nombre dentro de comillas, si existe crea una conexión.
cur=conn.cursor() ### ejecutar funciones  en BD
```

```python
#llamar bases de datos
data_general = pd.read_sql('SELECT * FROM data_general', con=conn)
employee_survey = pd.read_sql('SELECT * FROM employee_survey', con=conn)
manager_survey = pd.read_sql('SELECT * FROM manager_survey', con=conn)
retirement_info = pd.read_sql('SELECT * FROM retirement_info', con=conn)
```

```python
# para ver las dimensiones de las tablas y hacerme una idea de como deberian quedar las dimensiones despues de hacer la union
print(data_general.shape)
print(employee_survey.shape)
print(manager_survey.shape)
print(retirement_info.shape)
```

```python
# queria verificar si las fechas son las mismas
print(data_general['InfoDate'].unique())
print(employee_survey['DateSurvey'].unique())
print(manager_survey['SurveyDate'].unique())
#print(retirement_info['retirementDate'].unique())
```

```python
# Definir la consulta SQL para unir las tablas
consulta_sql = """
CREATE TABLE IF NOT EXISTS data_final AS
SELECT gd.*,
  es.EnvironmentSatisfaction, es.JobSatisfaction, es.WorkLifeBalance,
  ms.JobInvolvement, ms.PerformanceRating,
  ri.Attrition, ri.retirementDate, ri.retirementType, ri.resignationReason
FROM data_general AS gd
LEFT JOIN employee_survey AS es
ON gd.EmployeeID = es.EmployeeID AND strftime('%Y', gd.InfoDate) = strftime('%Y', es.DateSurvey)
LEFT JOIN manager_survey AS ms
ON gd.EmployeeID = ms.EmployeeID AND strftime('%Y', gd.InfoDate) = strftime('%Y', ms.SurveyDate)
LEFT JOIN (
  SELECT *
  FROM retirement_info
  WHERE strftime('%Y', retirementDate) = '2016'
) AS ri
ON gd.EmployeeID = ri.EmployeeID AND strftime('%Y', gd.InfoDate) = '2015';

CREATE TABLE IF NOT EXISTS datos_2015 AS
SELECT *
FROM data_final
WHERE strftime('%Y', InfoDate) = '2015'
AND (strftime('%Y', retirementDate) = '2016' OR retirementDate IS NULL);

-- Crear tabla datos_2016 con la información deseada
CREATE TABLE IF NOT EXISTS datos_2016 AS
SELECT *,
       NULL AS Attrition,
       NULL AS retirementDate,
       NULL AS retirementType,
       NULL AS resignationReason
FROM data_final
WHERE strftime('%Y', InfoDate) = '2016';
"""

# Confirmar los cambios y cerrar la conexión a la base de datos
conn.commit()
conn.close()
```
